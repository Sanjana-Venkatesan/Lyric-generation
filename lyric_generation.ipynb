```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("gpt2")

model = AutoModelForCausalLM.from_pretrained("gpt2")
```

```python
with open(pth, 'r', encoding='utf-8') as file:
    lyrics_content = file.read()
lines=lyrics_content.split('\n')
```

```python
len(lines)
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(lines,test_size=0.15)
len(train),len(test)
```

```python
with open('train.txt', 'w') as f:
  for t in train:
    f.write(t)
    f.write(' ')


with open('test.txt', 'w') as f:
  for t in test:
    f.write(t)
    f.write(' ')

train_path = 'train.txt'
test_path = 'test.txt'
```

```python
from transformers import TextDataset,DataCollatorForLanguageModeling

def load_dataset(train_path,test_path,tokenizer):
    train_dataset = TextDataset(
          tokenizer=tokenizer,
          file_path=train_path,
          block_size=128)

    test_dataset = TextDataset(
          tokenizer=tokenizer,
          file_path=test_path,
          block_size=128)

    data_collator = DataCollatorForLanguageModeling(
        tokenizer=tokenizer, mlm=False,
    )
    return train_dataset,test_dataset,data_collator

train_dataset,test_dataset,data_collator = load_dataset(train_path,test_path,tokenizer)
```

```python
from transformers import AutoModelWithLMHead, TrainingArguments
```

```python
from transformers import Trainer

training_args = TrainingArguments(
    output_dir="./gpt2-Taylor",
    overwrite_output_dir=True,
    num_train_epochs=30,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=64,
    eval_steps = 5,
    logging_steps=50,
    save_steps=0,
    warmup_steps=2,
    learning_rate = 2e-4
    )

trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=train_dataset,
    eval_dataset=test_dataset
)
```

```python
trainer.train()
```

```python
model_to_save = trainer.model.module if hasattr(trainer.model, 'module') else trainer.model  # Take care of distributed/parallel training
model_to_save.save_pretrained("outputs")
```
